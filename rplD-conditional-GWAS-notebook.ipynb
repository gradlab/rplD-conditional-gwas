### Metadata and phenotype filtering

In our previous GWAS study ([Ma and Mortimer et al., 2020](https://www.nature.com/articles/s41467-020-17980-1.pdf)), we compiled genomes and antibiotic resistance phenotypes for 4852 strains. We use this dataset ([Supplementary Data 3](files/metadata/MtrC-GWAS_Supplementary-Data-3_Global-Strain-Collection.csv)), and in particular the azithromycin MICs, for this study. We take a few filtering steps to clean up and standardize the MICs. Some isolates only had binary 'RES' or 'SUS' information; we removed those samples. Some isolates only had relative MIC values (e.g., >=8); we set those to be equal to the provided threshold (i.e., 8). We also log2-transformed MICs for both linear modeling and GWAS. Finally, we multiplied MICs of CDC GISP isolates prior to 2005 by two as previously done ([Grad et al., 2016](https://academic.oup.com/jid/article/214/10/1579/2290876); see also [Kersh et al., 2019](https://academic.oup.com/cid/article/70/5/798/5432326)) to account for a known change in the MIC measurement media protocol. We will use these corrected metadata files for the rest of the analysis.

```python
import pandas as pd
import numpy as np

metadata_filtered = pd.read_csv('metadata/MtrC-GWAS_Supplementary-Data-3_Global-Strain-Collection.csv')
metadata_filtered = metadata_filtered.replace('SUS', np.nan)
metadata_filtered = metadata_filtered.replace('RES', np.nan)

metadata_filtered['AZI_LOG'] = metadata_filtered['Azithromycin'].map(lambda x: float(str(x).lstrip('><=')))

adjusted_MIC = []
for index, row in metadata_filtered.iterrows():
    if not row['Reference'].startswith('Grad'):
        adjusted_MIC.append(row['AZI_LOG'])
    else:
        if int(row['Year']) < 2005:
            adjusted_MIC.append(2*row['AZI_LOG'])
        else:
            adjusted_MIC.append(row['AZI_LOG'])
            
metadata_filtered['Azithromycin'] = adjusted_MIC
metadata_filtered['AZI_LOG'] = np.log2(metadata_filtered['Azithromycin'])

metadata_filtered['Ceftriaxone'] = metadata_filtered['Ceftriaxone'].map(lambda x: float(str(x).lstrip('><=')))
metadata_filtered['CRO_LOG'] = np.log2(metadata_filtered['Ceftriaxone'])

metadata_filtered['Ciprofloxacin'] = metadata_filtered['Ciprofloxacin'].map(lambda x: float(str(x).lstrip('><=')))
metadata_filtered['CIP_LOG'] = np.log2(metadata_filtered['Ciprofloxacin'])

metadata_filtered.to_csv('metadata/strain-table-filtered.csv', index=False)
metadata_filtered.to_csv('metadata/gwas-strain-table-filtered.tsv', sep='\t', index=False, columns=['GWAS_ID', 'Country', 'Reference', 'AZI_LOG', 'CRO_LOG', 'CIP_LOG'])
```

## GWAS

We will now go through the code for running conditional GWAS in Pyseer. We also walk through some of our custom unitig annotation scripts and how to generate the log10 p-value vs. p-value plot in Figure 1, in the next section. 

To generate the unitigs file, we used [unitig-counter from Pyseer](https://github.com/johnlees/unitig-counter); we can provide our copy of the unitigs file on request. We controlled for population structure by converting the Gubbins recombination-corrected phylogeny to a similarity matrix available upon request. The shell scripts / paths for the GWAS runs are formatted for our SLURM cluster, but can be adapted for whatever cluster you're using. With 8 threads and 24GB memory total, Pyseer finishes running pretty quickly in about 30-40 minutes. We run GWAS conditional on the various covariates described below by supplying a tab-separated covariates file with a header ([metadata/combined-covariates.tsv](files/metadata/combined-covariates.tsv)) and specifying the covariate columns to be included using --use-covariates.

After conducting GWAS, we annotate unitigs using the following pipeline: first, we run Pyseer's annotate_hits_pyseer script, followed by a custom filtering / summarization script. This custom script ([software/scripts/summarize_top_unitigs.py](files/software/scripts/summarize_top_unitigs.py)) filters unitigs by length (short ones are discarded), allele frequency (major alleles are discarded), mapping location (unitigs that map to multiple spots in the genome are discarded), and annotation (similarly, unitigs mapping to known repetitive pilus / opacity protein genes are discarded). Unitigs have many advantages over traditional SNP-based GWASes, but also can introduce a lot of noise into the results; we've taken these stringent filtering steps to try to reduice that noise. We also then group unitigs iteratively that fall within a certain window (by default 25 bp on each side) of each other. This final step is similar to identifying and fine-mapping peaks within a Manhattan plot, and helps to group together redundant unitigs that encode the same genetic variant while preserving the resolution to detect multiple causal loci within one gene (e.g., we expect the two 23S rRNA mutations to appear as two different significant loci in the final output, rather than being grouped as one 23S rRNA result).

Example generic command:

```python
'''
python pyseer-runner.py                        # Depending on installation, can also call pyseer directly
    --phenotypes path/to/phenotypes/phenotypes.tsv
    --covariates path/to/covariates/covariates.tsv        
    --kmers path/to/unitigs/unitigs.txt 
    --similarity path/to/similarity/matrix/similarity.tsv
    
    --lmm 
    --phenotype-column DRUG 
    --output-patterns unitig_patterns.txt       # For calculating p-value threshold
    --use-covariates 1 2 3                      # Specify which covariates here 
    --uncompressed                              # If unitigs file is uncompressed
    
    > results.txt
'''
```

### Baseline GWAS (Supplementary Data 1)

### Linkage tables (Supplementary Data 1-3)

First, for each significant unitig, we need to extract the list of samples that have that unitig. This will be necessary for calculating linkage statistics in the plot. The code below parses the full unitigs file, but the results subsetted to the significant unitigs are located in [gwas/unitigs-samples/unitigs-samples.pkl](files/gwas/unitigs-samples/unitigs-samples.pkl), and hence this block does not need to be run if you're stepping through the notebook.

```python
import pandas as pd, pickle

unitigs = list(pd.read_csv('gwas/cond-dataset-23S/pyseer-AZI-cond-dataset-23S.results.significant.annotated.topkmers.txt', sep='\t')['Unitig'])
unitigs = unitigs + list(pd.read_csv('gwas/cond-dataset/pyseer-AZI-cond-dataset.results.significant.annotated.topkmers.txt', sep='\t')['Unitig'])
unitigs = unitigs + list(pd.read_csv('gwas/cond-23S/pyseer-AZI-cond-23S.results.significant.annotated.topkmers.txt', sep='\t')['Unitig'])
unitigs = unitigs + list(pd.read_csv('gwas/baseline/pyseer-AZI-baseline.results.significant.annotated.topkmers.txt', sep='\t')['Unitig'])
unitigs = set(unitigs)

unitigs_file = '/n/data1/hsph/immid/grad/gonococcus/analyses/unitigs/2019-04/unitigs/unitigs.txt'

unitig_samples = {}

for unitig in unitigs:
    with open(unitigs_file, 'r') as infile:
        for line in infile:
            if unitig == line.rstrip().split('|')[0].strip():
                samples = line.rstrip().split('|')[1].split()
                samples = [x.split(':')[0] for x in samples]
                unitig_samples[unitig] = samples
                break
                
with open('gwas/unitigs-samples/unitigs-samples.pkl', 'wb') as outfile:
    pickle.dump(unitig_samples, outfile)
```

Now, we calculate genetic linkage ($r^2$, the squared correlation coefficient) between significant unitigs and the 23S rRNA unitigs. Genetic linkage between two variants was defined as:

$$ r^2 = \frac{(p_{ij} - p_i p_j)^2 }{ p_i (1-p_i) p_j (1-p_j) } $$

where $p_i$ is the proportion of strains with variant $i$, $p_j$ is the proportion of strains with variant $j$, and $p_{ij}$ is the proportion of strains with both variants.

```python
import pickle

with open('gwas/unitigs-samples/unitigs-samples.pkl', 'rb') as infile:
    unitig_samples = pickle.load(infile)
```

```python
from math import sqrt

def calc_linkage(unitig, refunitig, unitig_samples):
    totaln = 4505
    samples = unitig_samples[unitig]
    p_x = len(samples) / totaln
    refsamples = unitig_samples[refunitig]
    p_23S = len(refsamples) / totaln
    p_both = len(list(set(samples) & set(refsamples))) / totaln
    D = p_both - p_x * p_23S
    r = D/sqrt(p_x*(1-p_x)*p_23S*(1-p_23S))
    return(r**2)
```

```python
a2059g_unitig = 'GCAAAGCTACAGTAAAGGTTCACGGGGTCTCTCCGTCTAGCAGCGGGTAGATTGCATCTTC'
c2611t_unitig = 'CGTCAAACTTCCAACGCCCACTGCAGATAGAGACCAAACTGTCTCACGACGTTTTAAACCC'
```

```python
# Add an r^2 column to annotated GWAS results and output as Supplementary Datasets

import pandas as pd

def output_datasets(gwas_results_file, dataset_file):
    gwas_results = pd.read_csv(gwas_results_file, dtype=object, sep='\t')
    r2 = []
    for unitig in list(gwas_results['Unitig']):
        r2.append(max(calc_linkage(unitig, a2059g_unitig, unitig_samples), calc_linkage(unitig, c2611t_unitig, unitig_samples)))
    gwas_results['r2'] = [round(x, 3) for x in r2]
    gwas_results.to_csv(dataset_file, sep='\t', index=False)

output_datasets('gwas/baseline/pyseer-AZI-baseline.results.significant.annotated.topkmers.txt', 'output/Supplementary-Data-1_GWAS.tsv')
output_datasets('gwas/cond-dataset-23S/pyseer-AZI-cond-dataset-23S.results.significant.annotated.topkmers.txt', 'output/Supplementary-Data-2_GWAS-cond-dataset-23S.tsv')
output_datasets('gwas/cond-dataset/pyseer-AZI-cond-dataset.results.significant.annotated.topkmers.txt', 'output/Supplementary-Data-3_GWAS-cond-dataset.tsv')
output_datasets('gwas/cond-23S/pyseer-AZI-cond-23S.results.significant.annotated.topkmers.txt', 'output/Supplementary-Data_GWAS-cond-23S.tsv')
```

```python
def proc_alignment(fname):
    samples = []
    with open(fname, 'r') as infile:
        for line in infile:
            if line.startswith('>'):
                sample = line[1:].rstrip()
                if sample.startswith('cds'):
                    continue
                samples.append(sample)
    return(samples)

all_samples = proc_alignment('genomic-epi/rplD-all.faa')
G70D_samples = proc_alignment('genomic-epi/rplD-G70D.faa')
other_samples = proc_alignment('genomic-epi/rplD-other.faa')
```

## Linear models

**We used an azithromycin MIC regression model based on known resistance determinants to show that inclusion of RplD variants significantly improved the model fit and reduced the number of strains with unexplained MIC variation.**

```python
%load_ext rpy2.ipython
```

```python
%%R -o azi_df -w 8 -h 4 --units in -r 400

library(tidyverse)
library(dplyr)
library(ggplot2)

azi_df = read.table("metadata/combined-covariates.csv", sep=",", header=TRUE, comment.char="", na.strings=c("", "NA"), stringsAsFactors=FALSE)
rplD = read.table("metadata/rplD-covariates.csv", sep=",", header=TRUE, comment.char="", na.strings=c("", "NA"), stringsAsFactors=FALSE)
pheno = read.table("metadata/strain-table-filtered.csv", sep=",", header=TRUE, comment.char="", na.strings=c("", "NA"), stringsAsFactors=FALSE)
pheno <- pheno %>%
    select(GWAS_ID, AZI_LOG, MtrC)

azi_df <- azi_df %>% 
    left_join(pheno, by = c("GWAS_ID")) %>% 
    left_join(rplD, by = c("GWAS_ID")) %>% 
    select(GWAS_ID, Country, AZI_LOG, MtrR_39, MtrR_45, mtr_promoter, 
           mtrR, Mosaic_Mtr_BAPS, rRNA23S_2059, rRNA23S_2611, MtrC, RplD_G70D, RplD_G68.A72_other_mutations) %>%
    filter(!is.na(AZI_LOG)) %>%

    replace(is.na(.), "Unmapped") %>% # Some mosaic alleles do not map well in SNP calling

    within(., Country <- relevel(factor(Country), ref = "USA")) %>% # Relevel baseline factors to faciliate interpretation
    within(., MtrR_39 <- relevel(factor(MtrR_39), ref = "A")) %>%
    within(., MtrR_45 <- relevel(factor(MtrR_45), ref = "G")) %>% 
    within(., mtr_promoter <- relevel(factor(mtr_promoter), ref = "A")) %>%
    within(., mtrR <- relevel(factor(mtrR), ref = "full_length")) %>%
    within(., Mosaic_Mtr_BAPS <- relevel(factor(Mosaic_Mtr_BAPS), ref = "11")) %>%
    within(., MtrC <- relevel(factor(MtrC), ref = "Intact"))

print(dim(azi_df))
head(azi_df, n=10)
```

```python
%%R -w 8 -h 4 --units in -r 400

library(jtools)
options(width = 250) 

model1 <- lm(AZI_LOG ~ (.-GWAS_ID-RplD_G70D-RplD_G68.A72_other_mutations), data=azi_df, x=TRUE, y=TRUE)
write.csv(tidy(model1) %>% select('term', 'std.error', 'estimate', 'p.value'), 'output/Supplementary-Data-4_Regression-model.csv', row.names = FALSE)
write.csv(tidy(model1) %>% select('term', 'std.error', 'estimate', 'p.value'), 'linear-models/lm-AZI.csv', row.names = FALSE)
summary(model1)

model2 <- lm(AZI_LOG ~ (.-GWAS_ID), data=azi_df, x=TRUE, y=TRUE)
write.csv(tidy(model2) %>% select('term', 'std.error', 'estimate', 'p.value'), 'output/Supplementary-Data-5_RplD-regression-model.csv', row.names = FALSE)
write.csv(tidy(model2) %>% select('term', 'std.error', 'estimate', 'p.value'), 'linear-models/lm-AZI-rplD.csv', row.names = FALSE)
summary(model2)
```

```python
%%R -w 8 -o residuals1 -o residuals2 -h 4 --units in -r 400

print(anova(model1, model2, test='Chisq'))

residuals1 <- resid(model1)
residuals2 <- resid(model2)
```

```python
total_n = len(residuals1)
model1_error = sum([True if abs(x) > 1 else False for x in residuals1])
model2_error = sum([True if abs(x) > 1 else False for x in residuals2])

print('''Out of {0} strains with azithromycin MICs, in the baseline model {1} strains ({2}%)
had MICs that deviated by more than one dilution from predicted MICs. After addition of RplD variants,
{3} strains ({4}%) had MICs that deviated by two dilutions or more from predicted MICs.
'''.format(total_n, model1_error, round(model1_error/total_n*100,2), model2_error, round(model2_error/total_n*100,2)))
```

### Variance explained by predictors (Supplementary Table 3)

To get a sense of the relative importance / fraction of variance explained by predictors, we use the relaimpo R package. A natural way to get a sense of the relative importance of a predictor is to see how much $R^2$ changes upon addition of the predictor. However, the order by which you add predictors matters, necessitating an arbitrary ordering of the predictors. The relaimpo package instead (when "lmg" is specified) calculates the change in $R^2$ for each predictor added averaged over all possible orderings of the model. The calculated metric sums to the total $R^2$ of the model.

```python
%%R

library(relaimpo)

azi_df_grouped <- azi_df %>% 
    mutate(RplD_any = ifelse((RplD_G70D == 1) | (RplD_G68.A72_other_mutations == 1), 1, 0)) %>%
    select(-RplD_G70D, -RplD_G68.A72_other_mutations)
gene.model <- lm(AZI_LOG ~ (.-GWAS_ID-Country), data=azi_df_grouped, x=TRUE, y=TRUE)
summary(gene.model)

calc.relimp(gene.model, type = c("first", "last", "lmg"))
```
